```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'D:\Ai\100-days-of-machine-learning-main\kmeans\student_clustering.csv')
dataset.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(7, 4))
plt.scatter(dataset['cgpa'], dataset['iq'], c='blue', alpha=0.5)
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('CGPA vs IQ Scatter Plot')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(dataset)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=4)
x= dataset.iloc[:,:].values
y_pred = km.fit_predict(x)
y_pred
```

```python
sns.scatterplot(x = dataset['cgpa'] , y = dataset['iq'], hue = y_pred , palette = 'bright')
```
